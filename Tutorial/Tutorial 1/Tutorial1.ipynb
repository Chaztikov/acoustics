## Solutions to the Wave Equation

Below is one solution of the wave equation for the pressure derived in the first lecture. 

\begin{equation}
p = p_A \cos\left(\omega t \pm kx \right)\quad \tag{1.1}
\end{equation}

$p_A$ is the amplitude, $k$ is the wave number and $\omega$ is the angular velocity. The sign is $+$ if the wave propagates in negative direction, and $-$ otherwise. Show that it is indeed a solution to the wave equation. Can you derive the relation between $k$ and $\omega$?

### Intuitively, what are $k$ and $\omega$?

$k$ represents the wavelength $\lambda$ of the wave:

\begin{equation}
k = \frac{2 \pi}{\lambda}
\end{equation}

$\omega$ represents the frequency $f$ or the period $T = \frac{1}{f}$ of the wave:

\begin{equation}
\omega = 2\pi f = \frac{2\pi}{T}
\end{equation}


### What about particle velocity versus speed of sound?


* The velocity of the particle's movement can be deduce from the linearized Euler equation

  \begin{equation}
  \rho_0\pad{u}{t}=-\pad{p}{x}\vb{e}_1
  \end{equation}

  After a short calculations we obtain the particle velocity $u(x,t)$

  \begin{equation}
  u(x,t)=\frac{p_A}{\rho_0\,c}\cos\left(\omega t - k x\right)
  \end{equation}



* The velocity of the wave itself, the phase velocity or most commonly termed the speed of sound is:

  \begin{equation}
  v_{\mathrm{sound}} = \frac{\lambda}{T}
  \end{equation}

  The speed of sound is depending on the environment as derived in the first lecture:

  \begin{equation}
  v_{\mathrm{sound}} = c = \frac{\cal B}{\rho_{0}}
  \end{equation}


### Now let's plot the wave

First we need to import neccessary libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt  # for plotting
import numpy as np  # for numeric operation
```

We then set the parameters: amplitude $p_A$, speed of sound, and frequency $f$, and from that we calculate wavelength $\lambda$, wave number $k$, and $\omega$.

```python
# Set the parameters
pA = 1. #Pa
c = 330. #m/s
f = 100. #Hertz

# Calculate lambda, k, and omega
wavelength=c/f
k = 2.*np.pi/wavelength
omega = 2.*np.pi*f
```

Good! Now we left with $x$ and $t$.
First let's create $x$, which will be an array that consists of many elements. Then for each element of $x$, we calculate the corresponding $y$ based on equation (1.1). Each set of $(x,y)$ forms a point on the graph, and connecting those points give us the shape of the wave.
Thus, the more point you have, the smoother the graph look like. Think of it as the 'resolution' of your graph.
For beginning, let's plot the wave at time $t=0$

```python
t = 0.
x = np.linspace(0, 10, 200) # this is a vector
p = pA * np.cos(omega*t - k*x)

# Plot the graph

plt.plot(x,p)
plt.xlabel('$x$')
plt.ylabel('$p_A$')
plt.title('Plotting the wave');
```

Below we demonstrate that the wave is actually moving to the right. For this we plot the wave at time $t_0$ and $t_1>t_0$.

```python
t0 = 0.
p0 = pA * np.cos(omega*t0 - k*x)

t1 = 0.2/f
p1 = pA * np.cos(omega*t1 - k*x)

# Plot the graph
fig = plt.figure(figsize=(8,6)) # to make a larger plot
plt.plot(x,p0,'-.',x,p1,'.')
plt.xlabel('$x$')
plt.ylabel('$p_A$')
plt.title('Travelling wave to the right');
```

Now you demonstrate the same for a wave travelling to the left.
