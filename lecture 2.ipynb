# Harmonic Plane Waves

In the following we will consider a constant speed of sound and start with the description of a plane wave propagation. Planar waves have a constant amplitude and phase perpendicular to the direction of propagation. Thus if the wave moves in the x-plane we have:

\begin{equation}
\ppad{p}{x}=\frac{1}{c^2}\ppad{p}{t}\quad .\tag{2.1}
\end{equation}

The complex solution to the harmonic wave equation is

\begin{equation}
p=A\,e^{i(\omega t - k x)}+B\,e^{i(\omega t + k x)}\quad . \tag{2.2}
\end{equation}

Using the linear Euler Eq. (1.11) we can obtain the particle velocity in a plane wave as

\begin{equation}
\vb{u}=u\,\va{e}_1=\left[\frac{A}{\rho_0 c}e^{i(\omega t - k x)}-\frac{B}{\rho_0 c}e^{i(\omega t + k x)}\right]\va{e}_1\tag{2.3}
\end{equation}


Please note that the constants $A$ and $B$, pressures and velocities are now all complex.
We can separate the wave travelling in positive $x$-direction from the the wave travelling in negative $x$-direction and designate them with the subscript $+$ and $-$, respectively.

The wave travelling in positive $x$-direction is

\begin{equation}
p_+=A e^{i(\omega t - k x)}\tag{2.4}
\end{equation}


while in negative $x$-direction is

\begin{equation}
p_-=B e^{i(\omega t + k x)\tag{2.5}}
\end{equation}


These two waves can be formally written for the other acoustic variables:

\begin{eqnarray}
u_{\pm}=\pm \frac{p_{\pm}}{\rho_0 c}\tag{2.6}\\
s_{\pm}=\frac{p_{\pm}}{\rho_0 c^2}\tag{2.7}
\end{eqnarray}

\begin{equation}
\Phi_{\pm}=-\frac{p_{\pm}}{i\omega\rho_0}\tag{2.8}
\end{equation}


Plane waves travelling in an arbitrary direction can be written as

\begin{equation}
p=A e^{i(\omega t-k_x x- k_y y - k_z z)}\tag{2.9}
\end{equation}


If we insert Eq. (2.9}) into the wave equation, Eq. (1.17) we obtain the condition

\begin{equation}
\left(\frac{\omega}{c}\right)^2 = k_x^2 +k_y^2+k_z^2\tag{2.10}\quad .
\end{equation}


Now we can introduce the propagation vector, $\vb{k}$ of the wave as

\begin{equation}
\vb{k}\equiv k_x \vb{e}_1 +k_y \vb{e}_2 +k_z \vb{e_3}\tag{2.11}
\end{equation}

with

\begin{equation}
\left| \vb{k} \right|=\frac{\omega}{c}\tag{2.12}\quad .
\end{equation}

Thus using the position vector $\vb{r}=x\vb{e}_1+y\vb{e}_2+z\vb{e}_3$ we can rewrite the plane wave solution, Eq. (2.9), as

\begin{equation}
p=A e^{i(\omega t-\vb{k}\vb{r})}\tag{2.13}\quad .
\end{equation}

The propagation vector points in the direction of wave propagation which can be simply shown. Surfaces of constant phase are defined by the scalar product $\vb{k}\cdot\vb{r}=\mathrm{const}$. The vector $\nabla (\vb{k}\cdot\vb{r})$ is a vector in the normal direction to the constant phase surface. But $\vb{k}=\nabla (\vb{k}\cdot\vb{r})$, thus $\vb{k}$ is pointing normal to the plane of constant phase and thus in the direction of propagation.

## Example

Let a plane wave travel in the xy-plane as sketched in [Fig. 2.1](#fig1). Formally the surface of constant phase is $\nabla(\vb{k}\cdot\vb{r})=0$; thus if the phase is independent of $z$, $\vb{k}$ can only have $k_x$ and $k_y$ components. We can thus write Eq. (2.9) as

\begin{equation}
p=A e^{i(\omega t-k_x x- k_y y)}\tag{2.14}
\end{equation}

and the equation for the surface of constant phase is

\begin{equation}
k_x x + k_y y = \mathrm{const.}\tag{2.15}
\end{equation}

or

\begin{equation}
y=-\frac{k_x}{k_y}x+\mathrm{const.}\tag{2.16}\quad .
\end{equation}

Equation (2.16) is a line with a slope $-k_x/k_y$. The surfaces are thus oriented as sketched in [Fig. 2.1](#fig1).

<img src="images/plane waves.png" width=450> <p>
<a id='fig1'><center> Figure 2.1 </center></a>

Now let's study this wave in the x-direction while $y=0$

\begin{equation}
p=A e^{i(\omega t-k_x x)}\tag{2.17}
\end{equation}

where $k_x=2\pi/\lambda_x$ with $\lambda_x$ being the apparent wavelength in $x$-direction. Note that this wavelength is larger than the wavelength in $\va{k}$-direction. We can relate $\lambda_x$ with the help of [Fig. 2.1](#fig1) to $\cos \phi=\lambda/\lambda_x$ and thus $k_x=k\cos\phi$. Applying the same arguments in the $y$-direction gives

\begin{equation}
\vb{k}=k \cos\phi \, \vb{e}_1 + k \sin\phi \, \vb{e}_2\tag{2.18}\quad .
\end{equation}

Thus an harmonic plane wave in 2-dimensions propagating under an angle $\phi$ to the $x$-axis can be written as

\begin{equation}
p=A e^{i(\omega t-k x \cos\phi - k y \sin\phi)}\tag{2.19}
\end{equation}


## Simulation of a 2-dimensional wave

The program solves the 2-dimensional wave equation. You can set the extend of the sound source, the duration of the sound and its wavelength. All boundaries are perfect reflectors thus the program simulates a perfect reveberant room. The blue dot is sensor of the acoustic pressure. The signal it receives is plotted as a function of time. 

```python
%matplotlib inline
import math as m
import numpy #array operations
import matplotlib.pyplot as plt #plotting
from ipywidgets import widgets #for the widgets
from IPython import display #for continous display

def plotwave(u,time,px,py,pp,pt):
    plt.clf()
    plt.subplot2grid((8,8),(0,0), colspan = 8, rowspan = 6)
    if w_scaled.value:
        plt.imshow(u, origin='upper', extent=[0., 2., 0., 2.]) #plot the wave field
    else:
        plt.imshow(u, origin='upper', extent=[0., 2., 0., 2.], vmax=2, vmin=-2) #plot the wave field
    
    plt.text(0.1,1.8,"time {0:.5f}".format(time)) #annotate the time
    plt.plot(w_probex.value,1.,'o') #position of probe
    plt.subplot2grid((8,8),(6,0), colspan = 8, rowspan = 2)
    plt.plot(pt,pp) #plot the pressure at the probe
    plt.gca().set_ylim([-2,2])
    display.clear_output(wait=True)
    display.display(plt.gcf())
        
def solvewave(b):
    tabs.visible=False
    #computational domain
    nx = ny = 128
    size=2. #size of the domain
    #parameters of the wave
    c = 5. #speed of sound
    l=w_wavelength.value #wavelength
    nu=c/l #frequency
    omega=nu*2.*m.pi #angular frequency
    duration=w_sourceduration.value/nu #duration of source
    #position
    emissionlength=(w_sizeemit.value/100.)*nx
    startx=int(nx/2-emissionlength/2)
    endx=int(nx/2+emissionlength/2)

    #further variables
    dx = size/(nx-1)
    CFL=0.1 #CFL number <1
    dt = CFL*dx/c 
    nt=int(w_simduration.value/dt) #number of time steps
    if w_position.value=='Left':
        sourcepos=0
    else:
        sourcepos=int(nx/2)

    #arrays for measuring the pressure at a position
    pt=numpy.arange(nt+1)*dt
    pp=numpy.zeros(nt+1)
    px=int(w_probex.value*(nx-1)/2.)
    py=int(ny/2)
        
    #every xx times over the total nt timesteps an output should be generated 
    output=map(int,list(numpy.linspace(1,nt,int(nt/50))))

    u  = numpy.zeros((nx,ny)) #pressure at t
    un = numpy.zeros((nx,ny)) #pressure at t-dt
    unn= numpy.zeros((nx,ny)) #pressure at t-2*dt

    row, col = u.shape
    #Assign initial conditions, for a sine wave
    un[startx:endx,0]=0.    #amplitude is sin(omega*t) with t=0
    unn[startx:endx,0]=1.   #velocity  is cos(omega*t) with t=0

    C=c*c*dt*dt/dx/dx

    plt.figure(1, figsize=(10, 10))

    for n in range(nt+1): ##loop across number of time steps
        #this line computes the finite differences of the wave equation
        u[1:-1,1:-1]=2.*un[1:-1,1:-1]-unn[1:-1,1:-1]+C*(un[1:-1,:-2]+\
                        un[:-2,1:-1]+un[2:,1:-1]+un[1:-1,2:]-4.*un[1:-1,1:-1])
        
        #hard reflective boundary conditions    
        u[0,:] = u[1,:]
        u[-1,:] = u[-2,:]
        u[:,0] = u[:,1]
        u[:,-1] = u[:,-2]
        
        #pressure source
        u[startx:endx,sourcepos]=m.sin(omega*n*dt)*float(n*dt<duration)
        
        pp[n]=u[py,px]
        
        #save values for the time derivative 
        unn=un.copy() #n-1 time steop
        un=u.copy()   #n time step
        
        if (n in output):
            plotwave(u,n*dt,px,py,pp,pt)
            
    #and plot the last figure    
    plotwave(u,n*dt,px,py,pp,pt)
    tabs.visible=True
    display.display(tabs)
    display.display(w_start)

#setup the graphical interface 
w_wavelength=widgets.FloatSlider(value=.5,min=0.1,max=1)
w_sizeemit=widgets.FloatSlider(value=50,min=1,max=100)
w_sourceduration=widgets.FloatSlider(value=1,min=0.5,max=10)
w_simduration=widgets.FloatSlider(value=.25,min=0.1,max=10)
w_scaled=widgets.Checkbox(description="Fixed Scale",value=False)
w_position=widgets.RadioButtons(options=["Left", "Center"],value="Center")
w_start=widgets.Button(description="Start Simulation")
w_start.on_click(solvewave)
w_probex=widgets.FloatSlider(description="Probe Position",value=0.5,min=0.,max=2.)
page1=widgets.VBox([widgets.HBox([widgets.HTML('Wavelength'),w_wavelength]),
                    widgets.HBox([widgets.HTML('Emitter Size in (%)'),w_sizeemit]),
                    widgets.HBox([widgets.HTML('Emission Duration in Periods'),w_sourceduration]),
                    widgets.HBox([widgets.HTML('Simulation Duration'),w_simduration]),
                    widgets.HBox([widgets.HTML('Source Position'),w_position])])
page2=widgets.Box([w_position])
page3=widgets.Box([w_scaled,w_probex])
tabs = widgets.Tab(children=[page1, page2, page3])
tabs.set_title(0, 'Wave')
tabs.set_title(1, 'Source')
tabs.set_title(2, 'Record')

display.display(tabs)
display.display(w_start)
```
