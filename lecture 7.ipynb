# Simple sources

An acoustic source is defined as the surface of a volume undergoing harmonic oscillations. The complex valued volume displacement is $V(t)$, which imposes a particle velocity on the fluid surrounding the source. The time derivative of $V(t)$ can be identified as the integral of its velocity over the surface. 

\begin{equation}
\pad{V}{t}=\int_S \vb{u} \cdot \vb{n} dS= Q \cdot e^{i\omega t}\quad .
\tag{7.1}
\end{equation}

Here $Q$ is terme dthe complex source strength. Then the pressure radiated from this source is (without derivation) given by 

\begin{equation}
p(r,t)=\frac{1}{2} i \rho_0 c \left(\frac{Q}{\lambda r}\right) e^{i(\omega t - k r)}\quad .
\tag{7.2}
\end{equation}

## Line source

The first example of simple sources besides the spherical shell is the line source. It consists of a cylinder with length $L$ and radius $a$ oscillating radially. The cylinder is aligned horizontally along the x-axis as shown in [Fig. 7.1](#fig1).

<img src="images/linesource.png" width=300> <p>
<a id='fig1'><center> Figure 7.1 </center></a>

The infinitesimal source strength $dQ$ is

\begin{equation}
dQ=U_0 2 \pi a \,\dd x
\tag{7.3}
\end{equation}

The complex pressure field emitted from the cylinder can be calculated by integrating over the infinitesimal sources, i.e. 

\begin{equation}
p(\vb{r},t)=\frac{1}{2} i \rho_0 c U_0 \frac{2 \pi a}{\lambda} \int_{-L/2}^{L/2} \frac{1}{r'} e^{i(\omega t - k r')} \dd x
\tag{7.4}
\end{equation}

It is important to note that the radiated sound pressure field depends only on $k a$ and not on the wavelength or the dimension of the cylinder alone. Thus let us write Eq. (7.4) again with $k a$:

\begin{equation}
p(\vb{r},t)=\frac{1}{2} i \rho_0 c U_0 k a \int_{-L/2}^{L/2} \frac{1}{r'} e^{i(\omega t - k r')} \dd x
\tag{7.5}
\end{equation}

The integral in Eq. (7.5) is difficult to solve as $r'$ depends complicated on $r$. Yet we can simplify Eq. (7.5) if we allow for distances $r\gg L$. That means that we are only describing the *far field* of the sound field.

Then the $\frac{1}{r'}\approx\frac{1}{r}$ and the $r'$ in the exp-function can be expressed as $r'\approx r-x\sin\Theta$ and we approximate Eq. (7.5) as:

\begin{equation}
p(r,\Theta,t)=\frac{1}{2} i \rho_0 c U_0 \frac{k a}{r} e^{i(\omega t -k r)} \int_{-L/2}^{L/2} e^{ik x\sin\Theta} \dd x
\tag{7.6}
\end{equation}

The integral can be evaluated and Eq. (7.6) becomes

\begin{equation}
p(r,\Theta,t)=\frac{1}{2} i \rho_0 c U_0 \frac{k a}{r} L \left(\frac{\sin(\frac{1}{2}k L \sin \Theta)}{\frac{1}{2}k L \sin \Theta}\right)e^{i(\omega t - k r)}
\tag{7.7}
\end{equation}


Let us now have a look of the pressure amplitude in the far field. The amplitude $P$ is a function of the distance $r$ and the angle $\Theta$. As the pressure amplitude must be a real positive value we can deduce it from Eq. (7.7):

\begin{equation}
P(r,\Theta)=\frac{1}{2} \rho_0 c U_0 \frac{k a}{r} L \left|\frac{\sin(\frac{1}{2}k L \sin \Theta)}{\frac{1}{2}k L \sin \Theta}\right|,
\tag{7.8}
\end{equation}

which can be split into a radial, $P_{ax}$ and angle or directional dependent pressure amplitude, $H(\Theta)$:

\begin{equation}
P(r,\Theta)=P_{ax}(r) H(\Theta)
\tag{7.9}
\end{equation}

with 

\begin{eqnarray}
P_{ax}=\frac{1}{2} \rho_0 c U_0 \frac{a}{r} k L\tag{7.10}\\
H(\Theta)=\left|\frac{\sin \theta}{\theta}\right|\tag{7.11}
\end{eqnarray}

where $\theta=\frac{1}{2} k L \sin \Theta$.

We see that the radial pressure drops with $r^{-1}$ in the far field. This is true for all simple sources.
The directional factor $H(\Theta)$ is a sinc function (or zeroth order Bessel function of the first kind).

The program below calculates and plots the directional factor.

```python
%matplotlib inline
import math as m
import numpy as np
import matplotlib.pyplot as plt

#calculate the dB scale
def dbnorm(x):
    return 20*np.log10(np.abs(x)/np.max(x))

theta=np.linspace(-m.pi/2.,m.pi/2.,1000)
kL=10
H=np.sin(.5*kL*np.sin(theta))/(.5*kL*np.sin(theta))
HdB=dbnorm(H)

plt.figure(figsize=(8,4), dpi=100)

ax1 = plt.subplot(121)

plt.plot(theta*180./m.pi,np.abs(H))
ax1.set_xlabel('angle (degree)')
ax1.set_ylabel('normalized amplitude')


ax2 = plt.subplot(122, projection='polar')

#to make the plot look nice
ax2.set_theta_zero_location('N');ax2.set_theta_direction('clockwise')
ax2.set_yticks(np.array([-12, -6, 0]));ax2.set_ylim(-20,0)
ax2.set_xticks(np.array([0, 45, 90, np.nan, np.nan, np.nan, 270, 315])/180.*m.pi)
ax2.set_xticklabels(('0','45','90','','','','-90','-45'))

c = plt.plot(theta, HdB)
plt.show()
```

## Plane Circular Piston

The next source is a baffeled piston. It is a disc of radius $a$ oscillating with a velocity $U_0$ and embedded in an infinite flat and rigid wall (baffle). The geometry is depicted in [Fig. 7.2](#fig2).

<img src="images/pistonsource.png" width=400> <p>
<a id='fig2'><center> Figure 7.2 </center></a>

The source strength is $dQ=U_0 dS$. Thus the radiated pressure can be obtained from the standard expression (the Rayleigh Integral)

\begin{equation}
p=\frac{1}{2} i \rho_0 c \frac{U_0}{\lambda} \int_S \frac{1}{r'} e^{i(\omega t - k r')} dS\quad.
\tag{7.12}
\end{equation}

Closed form solutions of Eq. (7.12) are only available for the axial pressure field ($\Theta=0$) and the far field approximation. We start with the axial pressure field

### Axial response

Along the axis the distance $r'$ is $\sqrt{r^2+\sigma^2}$ with $\sigma$ begin the radial position of the source, see [Fig. 7.2](#fig2). We obtain the following expression:

\begin{equation}
p=\frac{1}{2} i \rho_0 c \frac{U_0}{\lambda} e^{i\omega t} \int_0^a 
\frac{e^{-ik \sqrt{r^2+\sigma^2}}}{\sqrt{r^2 + \sigma^2}} 2 \pi \sigma d\sigma 
\tag{7.13}
\end{equation}

The kernel of Eq. (7.13) is a perfect differential and can therefore be integrated. We obtain:

\begin{equation}
p(r,0,t)=\rho_0 c U_0 \left[1-e^{-ik\left(\sqrt{r^2+a^2}-r\right)}\right] \cdot e^{i(\omega t - k r)}
\tag{7.14}
\end{equation}

The pressure amplitude, $P(r)$ is the magnitude of the real part of Eq. (7.14) which is

\begin{equation}
P(r,\theta=0)=2 \rho_0 c U_0 \left|\sin \frac{1}{2} k r \left(\sqrt{1+\left(\frac{a}{r}\right)^2}-1 \right)\right|
\tag{7.15}
\end{equation}


The pressure amplitude $P(r,\Theta=0)$, Eq. (7.15) is a rather complicated function with many extrema. We can simplify the expression if we look (1) at the far field $a \ll r$ and (2) demand that $\frac{r}{a}>\frac{ka}{2}$. The far field approximation allows to simplify the argument of the sin-function in Eq. (7.15):

\begin{equation}
P(r,\theta=0) \approx 2 \rho_0 c U_0 \left|\sin \frac{1}{2} k r \frac{1}{2} \left(\frac{a}{r}\right)^2 \right|=
2 \rho_0 c U_0 \left|\sin \frac{1}{2} \frac{ka}{2} \left(\frac{r}{a}\right)^{-1}\right|
\tag{7.16}
\end{equation}

The second approximation demands that we are sufficiently far from the source in terms of wavelengths (this becomes clearer below). Looking at the argument of the sin-function in the far field approximation Eq. (7.16) we see that we can then approximate $\sin(x)\approx x$ and obtain the expression

\begin{equation}
P_{ax} \approx \frac{1}{2} \rho_0 c U_0 \left(\frac{a}{r}\right) \left( k a\right)
\tag{7.17}
\end{equation}

Again we reveal the $r^{-1}$ dependency for simple sources in the far field approximation.

The second approximation leads directly to the Rayleigh length, $r>\frac{k a^2}{2}=\frac{\pi a^2}{\lambda}=\frac{S}{\lambda}$

But now let us study the Eq. (7.15) and find the position of the extrema

\begin{equation}
\frac{1}{2} k r \left(\sqrt{1+\left(\frac{a}{r}\right)^2}-1 \right) = \frac{m \pi}{2} \quad m=0,1,2,...
\tag{7.18}
\end{equation}

Simple rearrangement of Eq. (7.18) leads to the condition for extrema

\begin{equation}
\frac{r_m}{a}=\frac{a}{m\lambda} - \frac{m\lambda}{4 a}
\tag{7.19}
\end{equation}

Next we plot Eq. (7.15) as a function of $r/a$ with $k a=8 \pi$ and normalized by the maximum amplitude.

### Far field response
Now we want to calculate the emitted pressure field as a function of $\Theta$ under the far field approximation, $r \ll a$.
The geometry is depicted in [Fig. 7.3](#fig3).

<img src="images/piston2source.png" width=400> <p>
<a id='fig3'><center> Figure 7.3 </center></a>

We integrate in x-direction over line sources with length $2 a \sin \phi$, thus the source strength is

\begin{equation}
dQ=2 U_0 a \sin \phi dx
\tag{7.20}
\end{equation}

The infinitesimal pressure is

\begin{equation}
dp=2 \left( \frac{1}{2} i \rho_0 c \left( \frac{dQ}{\lambda r'}\right) e^{i(\omega t - k r')}\right)=
i\rho_0 c \frac{U_0}{\pi r'} k a \sin \phi e^{i(\omega t - k r')} dx
\tag{7.21}
\end{equation}

The factor 2 is the result of the baffle, effectivly doubling the emitted pressure. Please check Kinsler et al. (Fundamental of Acoustics, Chapter 7.2) for the derivation of the baffeled source.

In the far field $a \ll r$ we can approximate $r'$ in the exponent as $r'\approx r + \Delta r = r - a \sin\Theta \cos \phi$ and the $r'$ in the denominator of Eq. (7.21) becomes $r$ as done for the line source.

\begin{equation}
p(r,\Theta,t)=i \rho_0 c \frac{U_0}{\pi r} k a e^{i(\omega t - k r)}
\int_{-a}^a e^{i k a \sin \Theta cos \phi}\sin\phi \,\dd x
\tag{7.22}
\end{equation}

The integration can be replaced with $d\phi$, i.e. $dx=-a \sin \phi d\,\phi$ and we obtain:

\begin{equation}
p(r,\Theta,t)=i \rho_0 c \frac{U_0}{\pi} \frac{a}{r} k a e^{i(\omega t - k r)}
\int_0^\pi e^{i k a \sin \Theta cos \phi}\sin^2\phi \,\dd\phi
\tag{7.23}
\end{equation}


The integral is solvable leading to a Bessel function $J_1(z)$ with $z=k a \sin \Theta$.

\begin{equation}
p(r,\theta,t)=\frac{1}{2} i \rho_0 c U_0 \frac{a}{r} k a \left[ \frac{2 J_1(k a \sin \Theta)}{k a \sin \Theta}\right]
e^{i(\omega t - k r)}
\tag{7.24}
\end{equation}

The magnitude of the real pressure is

\begin{equation}
\left|p(r,\Theta)\right|=P_{ax}(r) H(\Theta)
\tag{7.25}
\end{equation}

where $P_{ax}$ is stated in Eq. (7.17). The directivity factor $H(\Theta)$ is 

\begin{equation}
H(\Theta)=\left|\frac{2 J_1 (z)}{z}\right| \quad \mathrm{with}\quad z= k a \sin \Theta
\tag{7.26}
\end{equation}

* Investigate the dependency of the pattern as a function of $k a$. What happens if you increase directivity?

```python
%matplotlib inline
import math as m
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp #for the Bessel function
#calculate the dB scale
def dbnorm(x):
    return 20*np.log10(np.abs(x)/np.max(x))

theta=np.linspace(-m.pi/2.,m.pi/2.,300)
ka=100./10.
H=2.*sp.jv(1,ka*np.sin(theta))/(ka*np.sin(theta))
HdB=dbnorm(H)

plt.figure(figsize=(8,8), dpi=100)

ax1 = plt.subplot(221)

plt.plot(theta*180./m.pi,np.abs(H))

ax2 = plt.subplot(222, projection='polar')

#to make the plot look nice
ax2.set_theta_zero_location('N');ax2.set_theta_direction('clockwise')
ax2.set_yticks(np.array([-30, -20, -10, 0]));ax2.set_ylim(-40,0)
ax2.set_xticks(np.array([0, 45, 90, np.nan, np.nan, np.nan, 270, 315])/180.*m.pi)
ax2.set_xticklabels(('0','45','90','','','','-90','-45'))

c = plt.plot(theta, HdB)
plt.show()
```

## Linear Array

The line array is a collection of $N$ simple sources separated each a distance $d$ apart. The geometry is depicted in [Fig. 7.4](#fig4). We first consider that all sources have the same strength and emit at the same phase. Then the j-th source emits a pressure wave of the form 

\begin{equation}
p(r,\Theta,t)=\sum_{j=1}^N \frac{A}{r'} e^{i(\omega t - k r')}
\tag{7.27}
\end{equation}

<img src="images/lineararray.png" width=400> <p>
<a id='fig4'><center> Figure 7.4 </center></a>

In the far field $L\ll r$, where $L=(N-1) d$ is the length of the array, we can assume that the radiated field is that of a superposition of plane waves from each source. From [Fig. 7.4](#fig4) we see that that phase difference between neighboring sources is $\Delta r= d \sin \Theta$. Also the path of the j-th source is that of the first source plus a phase $r'_i=r'_1-(j-1)\Delta r$. The path of the central ray at $L/2$ going to the $r$ can be related to the first source path length as $r=r'_1-\frac{L}{2d} \Delta r$. Again we make the approximation of $r\approx r'$ for the denominator in Eq. (7.27) but use the approximation for the inerference term in the exponent. We obtain

\begin{equation}
p(r,\theta,t)=\frac{A}{r}e^{-i \frac{L}{2d} k \Delta r} e^{i(\omega t -k r)} 
\sum_{j=1}^N e^{i(j-1)k \Delta r}
\tag{7.28}
\end{equation}


Using trigonometric identities we can rewrite Eq. (7.28) as

\begin{equation}
p(r,\theta,t)=\frac{A}{r} e^{i(\omega t -k r)} \left( \frac{\sin(\frac{N}{2}k \Delta r)}{\sin (\frac{1}{2} k \Delta r)}\right)
\tag{7.29}
\end{equation}

The pressure on the the axis where $\Theta=0$ is therefore

\begin{equation}
p(r,\theta,t)=N\frac{A}{r} e^{i(\omega t -k r)}
\tag{7.30}
\end{equation}

Thus the maximum pressure amplitude is 

\begin{equation}
P_{ax}=N\frac{A}{r} 
\tag{7.31}
\end{equation}

and the directivity factor

\begin{equation}
H(\Theta)=\left|\frac{1}{N} \frac{\sin(\frac{N}{2} k d \sin \Theta)}
{\sin(\frac{1}{2} k d \sin \Theta)}\right|
\tag{7.32}
\end{equation}

and we can write the far field pressure field in the convenient form 

\begin{equation}
P=P_{ax}\cdot H(\Theta)
\tag{7.33}
\end{equation}


### Major lobes

If $\frac{1}{2} k d |\sin \Theta|=m \pi$ with $m=0,1,2,...$ the denominator of Eq. (7.33) is 0, yet, also the nominator is 0 and we have maximum pressure from Eq. (7.31). Interestingly that allows for multiple maximum pressure lobes. Previously the maximum pressure was only radiated for $\Theta=0$. 

The angles for pressure maxima are

\begin{equation}
\left| \sin \Theta \right|= \frac{m \lambda}{d} \quad \quad m=0,1,2,... \frac{d}{\lambda}
\tag{7.34}
\end{equation}

Equation (7.34) can be rewritten as

\begin{equation}
d \left| \sin \Theta \right|= \Delta r = m \lambda \quad \quad m=0,1,2,... \frac{d}{\lambda}
\tag{7.35}
\end{equation}

which demonstrates that the maxima occur if the phase difference between neighboring emitters is a multiple of the wavelength.

<img src="images/lineararray2maxima.png" width=400> <p>
<a id='fig5'></a><center> Figure 7.5: Multiple pressure maxima for $kd=8$ and $N=8$. </center>

### Pressure zeros

The argument of the nominator sinus in Eq. (7.32) must be a multiple of $\pi$ while not leading to zero in the denominator. Then we have a pressure minimum.

\begin{equation}
\left| sin \Theta \right|=\frac{n \lambda}{N d} \quad \quad n=0,1,2,... \frac{N d}{\lambda} \quad \mathrm{and} \quad n\ne m N
\tag{7.36}
\end{equation}

### Single major lobe

It can be shown that for 

\begin{equation}
\frac{\lambda}{d}=\frac{N}{N-1}
\tag{7.37}
\end{equation}

a single major lobe is obtained which is almost as narrow as possible.

## Phased array

For arrays it is interesting to change the direction of the major lobes without physically rotating the array. This can be achieved by adding a constant delay (phase) between neighbouring sources, i.e. $j \tau$.

\begin{equation}
p(r,\Theta,t)=\sum_{j=1}^N \frac{A}{r'} e^{i(\omega (t+j\tau) - k r')}
\tag{7.38}
\end{equation}

The directional factor becomes

\begin{equation}
H(\Theta)=\left|\frac{1}{N} \frac{\sin\left(\frac{N}{2} k d (\sin \Theta-\sin \Theta_0)\right)}
{\sin(\frac{1}{2} k d \left(\sin \Theta-\sin \Theta_0)\right)}\right|
\tag{7.39}
\end{equation}


```python
%matplotlib inline
import math as m
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets

#calculate the dB scale
def dbnorm(x):
    return 20*np.log10(np.abs(x)/np.max(x))

#Plote the directivity factor as a function of kd and N
def pltlinarray(kd,N,sintheta0):
    theta=np.linspace(-m.pi/2.,m.pi/2.,500)

    H=1./N*np.sin(N/2.*kd*(np.sin(theta)-sintheta0))/np.sin(.5*kd*(np.sin(theta)-sintheta0))
    HdB=dbnorm(H)

    plt.figure(figsize=(8,8), dpi=100)

    ax1 = plt.subplot(221)

    plt.plot(theta*180./m.pi,np.abs(H))

    ax2 = plt.subplot(222, projection='polar')

    #to make the plot look nice
    ax2.set_theta_zero_location('N');ax2.set_theta_direction('clockwise')
    ax2.set_yticks(np.array([-30, -20, -10, 0]));ax2.set_ylim(-40,0)
    ax2.set_xticks(np.array([0, 45, 90, np.nan, np.nan, np.nan, 270, 315])/180.*m.pi)
    ax2.set_xticklabels(('0','45','90','','','','-90','-45'))

    c = plt.plot(theta, HdB)
    plt.show()

#Interactive plot
widgets.interact(pltlinarray,\
                 kd=widgets.FloatSlider(min=1.,max=20.,step=.25,value=8.),\
                 N=widgets.IntSlider(min=3,max=30,step=1,value=6),\
                sintheta0=widgets.FloatSlider(min=-.95,max=.95,step=.05,value=0.,\
                                              description=r'sin($\Theta_0$)'));
```
